##  Normal Distribution

Scenario ::
your portfolio return follow Normal distribution and you saw mean return of your portfolio is 1, and uts std is also 1 
then you thought what is the possiblity of you getting from -3% return to 5 percent return 

Importing libararies 

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Maine assume kiya ki meand aur std dono 1 ke brabr h 

```python
mean,std = 1,1
```

Maine -3 aur 5 ke beech me 250 points le liye h

```python
x = np.linspace(-3,5,250)
```

Abe ye saare 250 points h jo -3 aur 5 ke beech in sb ki probability density dekhte h 

```python
p = norm.pdf(x,1,1)
```

```python
plt.plot(x,p)
plt.title('Normal Distribution')
plt.xlabel('all 250 points')
plt.ylabel(' Probability Density of all points')
```

## Lognormal Distribution

Scenario:
Koi stock ka return log normla follow krta h aur tmne mean aur std liaya uska 1% ans 0.08%

```python
from scipy.stats import lognorm
```

mean aur std dono 5 percent diya hua h 

```python
mean,std = 0.02,0.05
```

Uska log mean and std nikal lete h 

```python
s = np.sqrt(np.log(1 + (std / mean)**2))
m = np.log(mean) - 0.5 * s**2
```

Hamne 0.02% se lekr 5% tk ka return dekhna h -> kon jyada frequently occur hota h 

```python
x = np.linspace(0.00002,0.05,100)
```

```python
log = lognorm.pdf(x,s = s,scale = np.exp(m))
```

```python
plt.plot(x, log)
plt.title('lognormal Distribution')
```

# Chi sq

```python
from scipy.stats import chi2
```

1. you expected 10% volatility of your portfolio
2. after checking last 12 month volatility you find 12%
you want to check is it normal or not 

```python
n,e,o = 12,0.1,0.12
```

yha pe maine chi nikal kr dekha kya ho rha h exactly 10 expected h aur 12 observed h

```python
chi = (n-1)*(o**2) / (e**2)
chi
```

Value aaya 15.83999 ab ye value dekhenge cumulative kitna percentage chances h

```python
p = 1 - chi2.cdf(chi,n-1)
p
```

 p 14.71% h ye jo area right side me hoga 

```python
x = np.linspace(0,30,100)
```

1. Ab maine kya kiya h ki 0 se lekr 30 puri value ko daal di aur hr value pe check krunga 
2. 10 expected tha aur 12 volatility me 15.8399 aaya 
3. lekin yha pr 0 se 30 tk saare chi values keliye check kr rha hu 

```python
pdf = chi2.pdf(x,n-1)
```

```python
plt.plot(x,pdf)
plt.axvline(x=15.8399, color='red', linestyle='--', linewidth=2, label='x = 15.8399')
plt.title('Chi sq Distribution')
```

Jaise aap dekh skte h chi = 15.83999 ke around 14.71% area right side pe h
