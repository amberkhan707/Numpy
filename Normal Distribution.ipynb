## Plotting Normal Distribution Curve 

Importing libararies 

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
min,max = -5,7
```

```python
x = np.linspace(min,max,250)
```

```python
p = norm.pdf(x,1,1)
```

```python
plt.plot(x,p)
```

## Poissons Distribution

```python
from scipy.stats import poisson
```

```python
pmf = poisson.pmf(np.arange(0,10),5)
```

```python
plt.plot(pmf)
```
